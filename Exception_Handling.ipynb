```python
#### Handle Division by Zero
def divisionerror(num):
    try:
        return num/0
    except ZeroDivisionError:
        print("Error: Number can not be divide  by 0.")

divisionerror(78)
```

```python
### Handle Invalid Input
def invalid_value(age):
    try:
        return int(age)

    except ValueError:
        print("Error: Invalid input, enter numbers only!!")
    except:
        print("Error: This variable is not created yet")

invalid_value("dhh")
```

```python
### Raise Custom Exception
def custom_exception():
    age=int(input("Enter your age: "))
    try:
        if (age<18):
            raise Exception("Error: Invalid age value!!!")
        return "Eligible!!!"
    except Exception as es:
        print(es)
custom_exception()
```

```python
### Handle List Index Error

def index_handling(lst,index):
    try:
        return lst[index]
    except IndexError:
        print("Error: this index no. is not present there in list")
    else :
        print(lst[index])
index_handling([23,45,1,2,3],4)
```
