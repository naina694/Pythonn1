```python
def bank_atm(amount):
    bal=10000
    try:
        if (amount<0):
            raise Exception("Negative amount entered, please enter valid amount")
        elif(amount>bal):
            raise Exception("Insufficient balance ")
    except ValueError:
        print("Please enter numbers only!!")
    except Exception as es:
        print(es)
    
    else:
        bal=bal-amount
        print("Remaining balance: ",bal)    
    finally:
        print("Balance: ",bal)
```

```python
bank_atm(int(input("Enter amount to wid: ")))
```
